```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
```

```python
df = pd.read_csv("weatherAUS.csv", sep=",")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
numeric_cols = ["MinTemp",
                "MaxTemp",
                "Rainfall",
                "Evaporation",
                "Sunshine",
                "WindGustSpeed",
                "WindSpeed9am",
                "WindSpeed3pm",
                "Humidity9am",
                "Humidity3pm",
                "Pressure9am",
                "Pressure3pm",
                "Cloud9am",
                "Cloud3pm",
                "Temp9am",
                "Temp3pm"]
df_num = df[numeric_cols].dropna()
ax, fig = plt.subplots(figsize=[16,9])
sns.heatmap(df_num.corr(),  annot=True)
```

Видим, что есть мультиколлениартные признаки, они могут помешать дальнейшему обучению моделей

```python
df = df.drop(["Pressure3pm","Temp9am","Temp3pm"], axis=1)
```

```python
df.isna().sum()
```

```python
numeric_cols.remove("Pressure3pm")
numeric_cols.remove("Temp9am")
numeric_cols.remove("Temp3pm")
```

```python
for col_name in numeric_cols:
  df[col_name] = df[col_name].fillna(df[col_name].mean())
```

```python
df.isna().sum()
```

```python
df["Date"] = df["Date"].astype("datetime64[ns]")
df["Day"] = df["Date"].dt.day
df["Month"]= df["Date"].dt.month
df["Year"] = df["Date"].dt.year
df = df.drop(["Date"], axis=1)
```

```python
df["Location"].unique()
```

```python
df["Location"].unique()
df = pd.get_dummies(data=df, prefix="Location", prefix_sep="_", columns=["Location"])
```

```python
df["WindGustDir"].unique()
```

```python
df["WindDir9am"].unique()
```

```python
df["WindDir3pm"].unique()
```

```python
cols = ["WindGustDir","WindDir9am","WindDir3pm"]
for col in cols:
  df[col] = df[col].fillna(df[col].mode())
df = pd.get_dummies(data=df, prefix=cols, prefix_sep="_", columns=cols)
```

```python
df["RainToday"].unique()
```

```python
df["RainTomorrow"].unique()
```

```python
cols_bool = ["RainToday", "RainTomorrow"]
for col in cols_bool:
  df[col] = df[col].map({'Yes': 1, 'No': 0})
  df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df["RainToday"].unique()
```

```python
df["RainTomorrow"].unique()
```

```python
df.info()
```

```python
df.to_csv("/content/drive/MyDrive/weather_AUS_preprocessed.csv", encoding="utf-8", sep=';', index=False)
```

Моделирование

```python
df_cv = df.sample(10000)
y_cv = df_cv['RainTomorrow']
X_cv = df_cv.drop(columns = ['RainTomorrow'])
y = df['RainTomorrow']
X = df.drop(columns = ['RainTomorrow'])
```

```python
from sklearn.model_selection import train_test_split
X_train_cv, X_test_cv, y_train_cv, y_test_cv = train_test_split(X_cv, y_cv, stratify=y_cv, test_size=0.2)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import confusion_matrix, classification_report
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_cv = scaler.fit_transform(X_train_cv)
X_test_cv = scaler.transform(X_test_cv)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
parameters = {'n_neighbors': [3, 5, 7, 9]}

knn_optimal = GridSearchCV(KNeighborsClassifier(), parameters).fit(X_train_cv, y_train_cv)
knn_optimal.best_params_
```

```python
knn = KNeighborsClassifier(knn_optimal.best_params_['n_neighbors']).fit(X_train, y_train)
```

```python
knn_pred = knn.predict(X_test)
```

```python
print(classification_report(y_test, knn_pred))
print(confusion_matrix(y_test, knn_pred))
```

```python
parameters = {'C': np.arange(0.1, 1, 0.1), 'penalty' : ['l1','l2']}
lr_optimal = GridSearchCV(LogisticRegression(solver='liblinear'), parameters).fit(X_train_cv, y_train_cv)
lr_optimal.best_params_
```

```python
lr = LogisticRegression(C = lr_optimal.best_params_['C'], penalty = lr_optimal.best_params_['penalty'], solver = 'liblinear').fit(X_train, y_train)
```

```python
lr_pred = lr.predict(X_test)
```

```python
print(classification_report(y_test, lr_pred))
print(confusion_matrix(y_test, lr_pred))
```

```python
parameters = {'max_depth': [3,5,7,9]}
dt_optimal = GridSearchCV(DecisionTreeClassifier(), parameters).fit(X_train_cv, y_train_cv)
dt_optimal.best_params_
```

```python
dt = DecisionTreeClassifier(max_depth = dt_optimal.best_params_['max_depth']).fit(X_train, y_train)
```

```python
dt_pred = dt.predict(X_test)
```

```python
print(classification_report(y_test, dt_pred))
print(confusion_matrix(y_test, dt_pred))
```

```python
parameters = {'n_estimators': [100, 200, 300, 400]}
bag_optimal = GridSearchCV(BaggingClassifier(estimator = DecisionTreeClassifier(max_depth = 5)), parameters).fit(X_train_cv, y_train_cv)
bag_optimal.best_params_
```

```python
bag = BaggingClassifier(estimator = DecisionTreeClassifier(max_depth = 5), n_estimators = bag_optimal.best_params_['n_estimators']).fit(X_train, y_train)
bag_pred = bag.predict(X_test)
```

```python
print(classification_report(y_test, bag_pred))
print(confusion_matrix(y_test, bag_pred))
```

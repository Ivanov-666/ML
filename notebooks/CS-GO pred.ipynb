```python
import pandas as pd
import numpy as np
df = pd.read_csv('../datasets/csgo_task.csv')
df
```

```python
df.info()
```

```python
df=df.fillna(df.median(numeric_only=True))
df.info()
```

```python
df = df.dropna()
df.info()
```

```python
df1 = pd.get_dummies(df.map, prefix='map')
df = df.drop(['map'], axis=1)
df = df.join(df1)
df
```

```python
df['bomb_planted']=df['bomb_planted'].values.astype(int)
df
```

```python
df.to_csv("../datasets/csgo_task_pred.csv")
```

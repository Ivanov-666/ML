```python
import numpy as np
import pandas as pd
from collections import defaultdict
import math
```

```python
class NaiveBayes1: 
    def __init__(self):
        pass
    def fit(self, X_train, y_train):
        self.classes_count, self.situations_count = defaultdict(lambda:0), defaultdict(lambda:0)
        for i in range(y_train.shape[0]):
            self.classes_count[y_train[i]]+=1
            for j in X_train[i]:
                self.situations_count[y_train[i], j]+=1
        for clas, sit in self.situations_count:
            self.situations_count[clas, sit] /= self.classes_count[clas]
        for clas in self.classes_count:
            self.classes_count[clas]/=y_train.shape[0]
    def predict(self,X_test):
        prediction = []
        for line in X_test:
            predict_ver = []
            for clas in self.classes_count:
                buffer = log(self.classes_count[clas])#log
                for predictor in line:
                    buffer+=log(self.situations_count.get((clas,predictor), 10**(-7))) #log
                predict_ver.append([clas,buffer])
            prediction.append(sorted(predict_ver, key=lambda x: x[1], reverse=True)[0][0])
        return prediction
```
